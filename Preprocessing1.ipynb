```python
# Step 1 : Import Required Libraries

import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
# Step 2 : Load Dataset

df = pd.read_csv("agriculture_pbl_dataset.csv")

# Displaying first 5 rows
df.head()
```

```python
# Step 3 : Checking Basic Information

# Check column names, datatype, missing values
df.info()
```

```python
# Step 4 : Remove Unnecessary Columns

# Region and Date are not useful for K-Means clustering
# because they are labels, not numerical learning features

df.drop(columns=["Region", "Date"], inplace=True)
```

```python
# Verifying columns after dropping
print(df.columns)
```

```python
# Step 5 : Convert Text Columns to Numeric

# Some columns contain values like:
# "8 hours", "75%", "120 kg", "7.2 pH"

cols_to_clean = [
    "Temperature",
    "Soil_pH",
    "Fertilizer",
    "Pesticide",
    "Sunlight",
    "Wind_Speed",
    "Crop_Diversity"
]

for col in cols_to_clean:
    
    # Extract only numeric values using regex
    df[col] = df[col].astype(str).str.extract(r'(\d+\.?\d*)')[0]
    
    # Convert extracted values into numeric datatype
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Checking if conversion worked
df.head()
```

```python
# Step 6 : Check Missing Values

# Display number of missing values in each column
print(df.isnull().sum())
```

```python
# Step 7 : Handle Missing Values

# Fill missing values using mean of each column
# Mean imputation is suitable for clustering projects
df.fillna(df.mean(numeric_only=True), inplace=True)

# Verify all missing values are removed
print(df.isnull().sum())
```

```python
# Step 8 : Check Duplicate Rows

# Count duplicate rows
print(df.duplicated().sum())
```

```python
# Step 9 : Remove Duplicate Rows

# Remove duplicates if any exist
# (Safe to run even if duplicates = 0)
df.drop_duplicates(inplace=True)
```

```python
# Step 10 : Handle Outliers using IQR Method

# Instead of deleting rows,
# we cap extreme values using clip()

for col in df.columns:
    
    # Calculate Q1 (25%) and Q3 (75%)
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    
    # Calculate IQR
    IQR = Q3 - Q1
    
    # Define lower and upper bounds
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    # Cap values outside bounds
    df[col] = df[col].clip(lower, upper)

#Displaying first 5 rows are outlier handling
print("Dataset after Outlier Handling:")
print(df.head())
```

```python
# Step 11 : Apply StandardScaler

# K-Means uses distance calculations
# so all features must be on same scale

scaler = StandardScaler()

# Fit and transform the dataset
scaled_data = scaler.fit_transform(df)

# Convert scaled array back to DataFrame
scaled_df = pd.DataFrame(scaled_data, columns=df.columns)

# Display first 5 rows after Standard Scaling
print("Dataset after Standard Scaling:")
print(scaled_df.head())
```

```python
# Step 12 : Verify Scaling

# Mean should be close to 0
# Standard deviation should be close to 1

print(scaled_df.describe())
```

```python
#Saving the cleaned dataset
scaled_df.to_csv("clean_scaled_agriculture_dataset.csv", index=False)
```
