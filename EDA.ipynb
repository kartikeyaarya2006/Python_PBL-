```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading dataset 
df = pd.read_csv('agriculture_eda_dataset.csv') 

# 1. Quick look at the first few rows
print("--- First 5 Rows ---")
display(df.head()) # Use print(df.head()) if not in a Jupyter Notebook

# 2. Check data types and look for missing values
print("\n--- Dataset Info ---")
df.info()

# 3. Get the math breakdown (mean, min, max, standard deviation)
print("\n--- Summary Statistics ---")
display(df.describe())

# 4. Count exactly how many missing values exist in each column
print("\n--- Missing Values Count ---")
print(df.isnull().sum())

```

```python
# Fill numeric with mean
df["Rainfall"].fillna(df["Rainfall"].mean(), inplace=True)

# Fill categorical with mode
df["Region"].fillna(df["Region"].mode()[0], inplace=True)
```

```python
# Histogram for multiple columns
df[["Rainfall","Temperature","Crop_Area","Avg_Yield"]].hist()

plt.title("Histogram of Key Features")
plt.show()


# Boxplot for selected columns
df[["Rainfall","Temperature","Crop_Area","Avg_Yield"]].boxplot()

plt.title("Boxplot for Outlier Detection")
plt.show()
```
