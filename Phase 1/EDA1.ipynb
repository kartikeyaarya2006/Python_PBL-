```python
#importing important libraries for EDA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#Loading the Dataset
df = pd.read_csv("agriculture_pbl_dataset.csv")
```

```python
#printing first 5 rows 
df.head()
```

```python
#getting the shape of the dataset
df.shape
```

```python
#getting the basic information of the dataset
df.info()
```

```python
#checking missing values of the dataset
df.isnull().sum()
```

```python
#viewing the missing values of the dataset in percentage
(df.isnull().sum() / len(df)) * 100
```

```python
#checking duplicate rows 
df.duplicated().sum()
```

```python
#Getting the statistical summary of the dataset
df.describe()
```

```python
df.select_dtypes(include=np.number)
```

```python
#Plotting Histograms 
df.hist(figsize=(18, 14), bins=20)
plt.tight_layout()
plt.show()
```

```python
#Using BoxPlot fot outlier detection
plt.figure(figsize=(18,10))
sns.boxplot(data=df.select_dtypes(include=np.number))
plt.xticks(rotation=90)
plt.show()
```

```python
"""Boxplot analysis revealed significant scale differences among features, 
particularly Crop Area, which contained very large values compared to other variables"""
plt.figure(figsize=(18,10))

numeric_df = df.select_dtypes(include=np.number).drop(columns=["Crop_Area"])

sns.boxplot(data=numeric_df)

plt.xticks(rotation=90)
plt.show()
```

```python
#Getting the Correlation Heatmap
plt.figure(figsize=(16,12))
sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap='coolwarm',
            fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Scatterplot for Rainfall Vs Average Yield
sns.scatterplot(x='Rainfall', y='Avg_Yield', data=df)
plt.title("Rainfall vs Average Yield")
plt.show()
```

```python
#Scatterplot for Irrigation Vs Average Yield
sns.scatterplot(x='Irrigation', y='Avg_Yield', data=df)
plt.title("Irrigation vs Average Yield")
plt.show()
```

```python
#Scatterplot for Nitrogen Vs Average Yield
sns.scatterplot(x='Nitrogen', y='Avg_Yield', data=df)
plt.title("Nitrogen vs Average Yield")
plt.show()
```
