```python
#Importing the cleaned dataset
import pandas as pd

scaled_df = pd.read_csv("clean_scaled_agriculture_dataset.csv")

scaled_df.head()
```

```python
# Using Backward Elimination (OLS Method)
```

```python
#Step 1: Importing important libraries for OLS regression
import statsmodels.api as sm
```

```python
# Step 2 : Define Features (X) and Target (y)

# We use Avg_Yield as proxy target
# because K-Means is unsupervised and has no label

# X = all features except Avg_Yield
X = scaled_df.drop("Avg_Yield", axis=1)

# y = target column
y = scaled_df["Avg_Yield"]


# Step 3 : Add Constant

# OLS requires adding a constant (intercept)

X = sm.add_constant(X)


# Step 4 : Train OLS Model

# Fit Ordinary Least Squares regression model

model = sm.OLS(y, X).fit()


# Step 5 : Display Full Summary

# This shows:
# - p-values
# - R-squared
# - feature significance

print(model.summary())


# BACKWARD ELIMINATION PROCESS

# Rule:
# If p-value > 0.05 → remove feature
# If p-value < 0.05 → keep feature

# We remove the feature with highest p-value first,
# then run the model again


# Step 6 : Check P-values Only

print(model.pvalues)


# Step 7 : Remove Weak Feature Manually

# Example:
# Suppose Potassium has highest p-value > 0.05

# Replace "Potassium" with actual weak feature
# after checking model.summary()

X = X.drop("Potassium", axis=1)


# Step 8 : Rebuild Model After Removal

model = sm.OLS(y, X).fit()

print(model.summary())


# Step 9 : Repeat Until All p-values < 0.05

# Continue removing highest p-value feature
# one by one until all remaining features are significant


# Step 10 : Final Selected Features

print("Final Selected Features:")
print(X.columns)
```

```python
# =========================================
# END OF FEATURE SELECTION
# =========================================
```

```python
# Features with p-value > 0.05 are considered
# statistically weak contributors toward Avg_Yield.

# Such features may introduce unnecessary noise
# and reduce clustering quality in K-Means.

# Therefore, we remove the feature with the
# highest p-value first and rebuild the model.

# Example:
# Groundwater has a very high p-value (> 0.05),
# which means it has weak contribution toward
# predicting Avg_Yield, so it is removed first.

X = X.drop("Groundwater", axis=1)
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
# Humidity has a very high p-value (> 0.05),
# which means it has weak contribution toward
# predicting Avg_Yield, so it is removed.

X = X.drop("Humidity", axis=1)
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
# Altitude has a very high p-value (> 0.05),
# which means it has weak contribution toward
# predicting Avg_Yield, so it is removed.

X = X.drop("Altitude", axis=1)
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
# Wind Speed has a very high p-value (> 0.05),
# which means it has weak contribution toward
# predicting Avg_Yield, so it is removed.

X = X.drop("Wind_Speed", axis=1)
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
#Saving the final dataset after feature selection 

final_selected_df = X.drop("const", axis=1)

final_selected_df.to_csv(
    "final_selected_features_dataset.csv",
    index=False
)

print("Final Selected Dataset Saved Successfully ✅")
```
