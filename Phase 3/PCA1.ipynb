```python
# PHASE 5 : PRE-CLUSTERING VISUALIZATION
# Using PCA (Principal Component Analysis)
```

```python
# Step 1 : Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.decomposition import PCA
```

```python
# Step 2 : Load Final Selected Dataset

# Load the dataset after feature selection
final_df = pd.read_csv("final_selected_features_dataset.csv")

# Display first 5 rows
print(final_df.head())
```

```python
# Step 3 : Apply PCA

# PCA reduces high-dimensional data
# into lower dimensions for visualization
# Here:
# 14 features → 2 Principal Components

pca = PCA(n_components=2)

pca_data = pca.fit_transform(final_df)
```

```python
# Step 4 : Convert PCA Output to DataFrame
pca_df = pd.DataFrame(
    data=pca_data,
    columns=["PC1", "PC2"]
)

print(pca_df.head())
```

```python
# Step 5 : Check Explained Variance

# This tells how much information
# is preserved in PC1 and PC2

print("Explained Variance Ratio:")
print(pca.explained_variance_ratio_)
```

```python
# Step 6 : PCA Scatter Plot

# Visualize the data before clustering

plt.figure(figsize=(10, 6))

sns.scatterplot(
    x="PC1",
    y="PC2",
    data=pca_df
)

plt.title("PCA Visualization Before Clustering")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")

plt.show()
```

```python
# Step 7 : Save PCA Dataset

pca_df.to_csv("pca_visualization_dataset.csv", index=False)

print("PCA Visualization Completed Successfully")
```
